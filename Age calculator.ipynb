```python
# Age calculator
from tkinter import *
from datetime import date

today=date.today()

def button_clear():
    e1.delete(0, END)
    e2.delete(0, END)
    e3.delete(0, END)
    e4.delete(0, END)

def get_age():
    d= int(e1.get())
    m=int(e2.get())
    y=int(e3.get())
    age = today.year-y-((today.month, today.day)<(m,d))
    e4.insert(10,str(age))

root=Tk()
root.resizable(0,0)
root.title("Age Calculator")
root.geometry('350x330')


label1=Label(root,text='Age Calculator!',font=(40),width=30,fg="red")
label2=Label(root,text='  Enter your Date of Birth:',width=20,font=(8))
label3=Label(root,text='Date:')
label4=Label(root,text='Month:')
label5=Label(root,text='Year:')
label6=Label(root,text="\nYour Age:",font=(12))
l=Label(root,text=' ',font=(20))

e1=Entry(root,font=(4))
e2=Entry(root,font=(4))
e3=Entry(root,font=(4))
e4=Entry(root,font=(4))

button1=Button(root,text='Calculate Age',fg='blue',width=15,font=(1),command=get_age)
button2=Button(root,text='  Clear  ',font=(18),bg='lavender',command=button_clear)

label1.grid(row=0,column=0,columnspan=4)
label2.grid(row=1,column=0,columnspan=4)
label3.grid(row=2,column=1)
label4.grid(row=3,column=1)
label5.grid(row=4,column=1)
label6.grid(row=6,column=0,columnspan=3)
l.grid(row=9,column=0)

e1.grid(row=2,column=2)
e2.grid(row=3,column=2)
e3.grid(row=4,column=2)
e4.grid(row=7,column=1,columnspan=2)

button1.grid(row=5,column=0,columnspan=4)
button2.grid(row=10,column=0,columnspan=4)

root.mainloop()
```
