```python
#Tic-Toc-Toi
import random

def display(board):
    for i in board:
        for j in i:
            print(j,end=' ')
        print()
        
def winner(board,player):
    
    #horizontal
    for i in range(3):
        flag=1
        for j in range(3):
            if board[i][j]!=player:
                flag=0
                break
        if flag:
            return flag
    
    #vertical
    for i in range(3):
        flag=1
        for j in range(3):
            if board[j][i]!=player:
                flag=0
                break
        if flag:
            return flag
        
    #primary diagnol
    for i in range(3):
        flag=1
        if board[i][i]!=player:
            flag=0
            break
    if flag:
        return flag
    
    #secondary daigonal
    for i in range(3):
        flag=1
        if board[i][i]!=player:
            flag=0
            break
    if flag:
        return flag
    
    flag=0
    return flag


def board_full(board):
    for i in board:
        for j in i:
            if j == '-':
                 return False
    return True

        
def create_board(board):
    for i in range(3):
        row = []
        for j in range(3):
            row.append('-')
        board.append(row)
        
def start(board):
    num=random.randint(0,1)
    player='X' if num==1 else 'O'
    while True:
        print()
        print(player," Turn :")
        display(board)
        row, col = list(map(int, input("Enter row and column numbers to fix spot: ").split()))
        board[row-1][col-1]=player
        if winner(board,player):
            print()
            display(board)
            print(player," won the game..!")
            break
        if board_full(board):
            print()
            display(board)
            print("Match Tie..!")
            break
        player='X' if player=='O' else 'O'
        display(board)
    
    


while True:
    board=[]
    print("\n\n***TIC-TOC-TOI***")
    create_board(board)
    display(board)
    print("Lets Start.!")
    start(board)
    
    s=input("\nDo you want to play again?(yes/no):")
    if s!='yes':
        print("Thank You. :)")
        break
    else:
        player='X' if player=='O' else 'O'
    
```
